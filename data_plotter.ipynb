```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import font_manager
import pandas as pd
import datetime
```

```python
dColours = plt.rcParams['axes.prop_cycle'].by_key()['color']
label_font = font_manager.FontProperties(
    family=['Fira Sans'],
    weight='regular',
    size=20
)
tick_font = font_manager.FontProperties(
    family=['Fira Sans'],
    weight='regular',
    size=14
)
```

```python
the_data = pd.read_csv("../samoa/samoa-data/samoa_press_release_data.csv", header=0, dayfirst=True, parse_dates=True, skiprows=[1])
the_dates = list(map(lambda t: datetime.datetime.strptime(t, "%d/%m/%Y"), the_data.Date.values))
raw_times, raw_cumdata, H, cH, D, G = the_data.iloc[:,1:].values.T
```

```python
data = raw_cumdata[0:None]
data_early = data[:15]
data_late = data[15:]
```

```python
plt.figure(figsize=[13,8],dpi=360)

plt.plot(the_dates, data, 'o', color='darkred')
plt.xticks(font_properties=tick_font)
plt.yticks(font_properties=tick_font)
plt.xlabel("Date", font_properties=label_font)
plt.title("Cumulative Reported Cases", font_properties=label_font)
plt.box(False)
plt.grid(True)
```

```python
plt.figure(figsize=[13,8],dpi=360)

plt.plot(the_dates[:15], data_early, 'o', color='darkred')
plt.plot(the_dates[15:], data_late, 'o', color='red', alpha=0.2)
plt.xticks(font_properties=tick_font)
plt.yticks(font_properties=tick_font)
plt.xlabel("Date", font_properties=label_font)
plt.title("Cumulative Reported Cases", font_properties=label_font)
plt.box(False)
plt.grid(True)
```

```python
plt.figure(figsize=[13,8],dpi=360)

plt.plot(the_dates, np.gradient(data, raw_times), 'o', color='darkred')
plt.xticks(font_properties=tick_font)
plt.yticks(font_properties=tick_font)
plt.xlabel("Date", font_properties=label_font)
plt.title("Report Incidence", font_properties=label_font)
plt.box(False)
plt.grid(True)
```

```python
plt.figure(figsize=[13,8],dpi=360)

plt.plot(the_dates, D, 'o', color='darkred')
plt.xticks(font_properties=tick_font)
plt.yticks(font_properties=tick_font)
plt.xlabel("Date", font_properties=label_font)
plt.title("Cumulative Deaths", font_properties=label_font)
plt.box(False)
plt.grid(True)
```

```python
plt.figure(figsize=[13,8],dpi=360)

plt.plot(the_dates, H, 'o', color='darkred')
plt.xticks(font_properties=tick_font)
plt.yticks(font_properties=tick_font)
plt.xlabel("Date", font_properties=label_font)
plt.title("Hospitalisations", font_properties=label_font)
plt.box(False)
plt.grid(True)
```
