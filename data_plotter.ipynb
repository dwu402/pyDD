```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from matplotlib import font_manager
import pandas as pd
import datetime
```

```python
dColours = plt.rcParams['axes.prop_cycle'].by_key()['color']
label_font = font_manager.FontProperties(
    family=['Fira Sans'],
    weight='regular',
    size=20
)
tick_font = font_manager.FontProperties(
    family=['Fira Sans'],
    weight='regular',
    size=14
)
pub_font = font_manager.FontProperties(
#     family=['cmr10'],
    weight='regular',
    size=32
)
publish = True
if publish:
    matplotlib.rcParams.update(
        {
            'text.usetex': False,
            'font.family': 'cmr10',
            'mathtext.fontset': 'cm',
        }
    )
```

```python
the_data = pd.read_csv("../samoa/samoa-data/samoa_press_release_data.csv", header=0, dayfirst=True, parse_dates=True, skiprows=[1])
the_dates = list(map(lambda t: datetime.datetime.strptime(t, "%d/%m/%Y"), the_data.Date.values))
raw_times, raw_cumdata, H, cH, D, G = the_data.iloc[:,1:].values.T
```

```python
data = raw_cumdata[0:None]
data_early = data[:15]
data_late = data[15:]
```

```python
date_tick = [datetime.date(2019, M, 1) for M in [10, 11, 12]] + [datetime.date(2020, 1, 1)]
date_tick_label = ['Oct\n2019', 'Nov', 'Dec', 'Jan\n2020']
```

```python
plt.figure(figsize=[13,8],dpi=360)

plt.plot(the_dates, data, 'o', color='k')
plt.xticks(date_tick, date_tick_label, font_properties=pub_font)
plt.yticks(np.arange(0,6500,2000), font_properties=pub_font)
# plt.xlabel("Date", font_properties=label_font)
# plt.title("Cumulative Reported Cases", font_properties=label_font)
plt.gca().yaxis.set_ticks_position('both')
plt.ylim([0,6000])
plt.box(True)
plt.grid(False)
```

```python
plt.figure(figsize=[13,8],dpi=360)

plt.plot(the_dates[:15], data_early, 'o', color='darkred')
plt.plot(the_dates[15:], data_late, 'o', color='red', alpha=0.2)
plt.xticks(font_properties=tick_font)
plt.yticks(font_properties=tick_font)
plt.xlabel("Date", font_properties=label_font)
plt.title("Cumulative Reported Cases", font_properties=label_font)
plt.box(False)
plt.grid(True)
```

```python
plt.figure(figsize=[13,8],dpi=360)

plt.plot(the_dates, np.gradient(data, raw_times), 'o', color='k')
plt.xticks(date_tick, date_tick_label, font_properties=pub_font)
plt.yticks([0,100,200,300],font_properties=pub_font)
# plt.xlabel("Date", font_properties=label_font)
# plt.title("Report Incidence", font_properties=label_font)
plt.gca().yaxis.set_ticks_position('both')
plt.ylim([0,300])
plt.box(True)
plt.grid(False)
```

```python
plt.figure(figsize=[13,8],dpi=360)

plt.plot(the_dates, D, 'o', color='k')
plt.xticks(date_tick, date_tick_label, font_properties=pub_font)
plt.yticks([0,50,100],font_properties=pub_font)
# plt.xlabel("Date", font_properties=label_font)
# plt.title("Cumulative Deaths", font_properties=label_font)
plt.gca().yaxis.set_ticks_position('both')
plt.ylim([0,100])
plt.box(True)
plt.grid(False)
```

```python
plt.figure(figsize=[13,8],dpi=360)

plt.plot(the_dates, H, 'o', color='darkred')
plt.xticks(font_properties=tick_font)
plt.yticks([0,50,100,150,200],[0,None,None,None,200],font_properties=tick_font)
# plt.xlabel("Date", font_properties=label_font)
# plt.title("Hospitalisations", font_properties=label_font)
plt.ylim([0,200])
plt.box(True)
plt.grid(False)
```

```python
plt.figure(figsize=[13,8],dpi=360)

plt.plot(the_dates, cH, 'o', color='k')
plt.xticks(date_tick, date_tick_label, font_properties=pub_font)
plt.yticks([0,1000,2000],font_properties=pub_font)
# plt.xlabel("Date", font_properties=label_font)
# plt.title("Cumulative Hospitalisations", font_properties=label_font)
plt.gca().yaxis.set_ticks_position('both')
plt.ylim([0,2000])
plt.box(True)
plt.grid(False)
```
